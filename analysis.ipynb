## 1. Setup – Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import seaborn as sns
from scipy import stats

# Notebook display settings
pd.set_option('display.max_columns', 20)
pd.set_option('display.float_format', '{:.2f}'.format)
plt.rcParams.update({'figure.figsize': (10, 5), 'axes.titlesize': 14})
sns.set_theme(style='whitegrid')

print('Libraries loaded successfully.')
```

## 2. Load Dataset

We use the well-known **Iris** dataset (built into seaborn) as a demonstration dataset.
Replace the cell below with your own `pd.read_csv(...)` or equivalent loader when using a custom dataset.

```python
df = sns.load_dataset('iris')

print(f'Dataset shape: {df.shape}')
df.head()
```

## 3. Data Inspection

```python
print('Column types:')
print(df.dtypes)
print()
print('Missing values per column:')
print(df.isnull().sum())
```

```python
df.describe()
```

## 4. Exploratory Data Analysis (EDA)

```python
# Class distribution
species_counts = df['species'].value_counts()

fig, ax = plt.subplots()
species_counts.plot(kind='bar', ax=ax, color=['steelblue', 'seagreen', 'tomato'], edgecolor='white')
ax.set_title('Sample Count per Species')
ax.set_xlabel('Species')
ax.set_ylabel('Count')
ax.yaxis.set_major_locator(mticker.MaxNLocator(integer=True))
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# Distribution of numeric features
numeric_cols = df.select_dtypes(include='number').columns.tolist()

fig, axes = plt.subplots(1, len(numeric_cols), figsize=(16, 4))
for ax, col in zip(axes, numeric_cols):
    df[col].hist(ax=ax, bins=20, color='steelblue', edgecolor='white')
    ax.set_title(col.replace('_', ' ').title())
    ax.set_xlabel('Value')
    ax.set_ylabel('Frequency')
plt.suptitle('Feature Distributions', y=1.02)
plt.tight_layout()
plt.show()
```

```python
# Box plots by species
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
axes = axes.flatten()
colors = ['steelblue', 'seagreen', 'tomato']

for ax, col in zip(axes, numeric_cols):
    df.boxplot(column=col, by='species', ax=ax,
               boxprops=dict(color='dimgray'),
               medianprops=dict(color='crimson', linewidth=2))
    ax.set_title(col.replace('_', ' ').title())
    ax.set_xlabel('')

plt.suptitle('Feature Distribution by Species', y=1.02)
plt.tight_layout()
plt.show()
```

## 5. Correlation Analysis

```python
corr = df[numeric_cols].corr()

fig, ax = plt.subplots(figsize=(7, 5))
sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm',
            vmin=-1, vmax=1, linewidths=0.5, ax=ax)
ax.set_title('Pearson Correlation Matrix')
plt.tight_layout()
plt.show()
```

```python
# Scatter plot – petal length vs petal width coloured by species
fig, ax = plt.subplots()
for species, grp in df.groupby('species'):
    ax.scatter(grp['petal_length'], grp['petal_width'],
               label=species, alpha=0.7, edgecolors='white', s=60)

ax.set_xlabel('Petal Length (cm)')
ax.set_ylabel('Petal Width (cm)')
ax.set_title('Petal Length vs Petal Width by Species')
ax.legend(title='Species')
plt.tight_layout()
plt.show()
```

## 6. Statistical Testing

Test whether the mean **petal length** differs between species using a one-way ANOVA.

```python
groups = [grp['petal_length'].values for _, grp in df.groupby('species')]
f_stat, p_value = stats.f_oneway(*groups)

print(f'One-way ANOVA – Petal Length across species')
print(f'  F-statistic : {f_stat:.2f}')
print(f'  p-value     : {p_value:.2e}')

alpha = 0.05
if p_value < alpha:
    print(f'\nConclusion: Reject H₀ (p < {alpha}). '
          'Mean petal lengths differ significantly between species.')
else:
    print(f'\nConclusion: Fail to reject H₀ (p ≥ {alpha}). '
          'No significant difference found.')
```

## 7. Pair Plot Overview

```python
pair_grid = sns.pairplot(df, hue='species', diag_kind='kde',
                         plot_kws={'alpha': 0.6, 's': 40})
pair_grid.figure.suptitle('Pair Plot of All Features', y=1.02)
plt.show()
```
